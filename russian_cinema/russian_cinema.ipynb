#### Шаг 2.1. Проверьте типы данных

- Проверьте типы данных в датафрейме и преобразуйте их там, где это необходимо.

При обзоре данных мы заметили, что необходимо поменять тип данных в столбцах `show_start_date` на datetime, `ratings` на float.

```python
#Меняем тип данных в столбце `show_start_date` на datetime
data['show_start_date'] = pd.to_datetime(data['show_start_date'], format = '%Y-%m-%d')
```

Прежде чем менять тип данных в столбце `ratings`, необходимо найти значения, из-за которых тип данных object.

```python
#Выведем все значения столбца `show_start_date`
data['ratings'].unique()
```

Тут есть значения в %, хотя рейтинг в Кинопоиске указывается в баллах. В процентах есть некий рейтинг ожидания, который не имеет никакого отношения к рейтингу фильма. От этих значений необходимо избавиться. Но удалять эти строки мы не будем, так как там содержатся другие важные данные, а рейтинг - лишь один параметр для анализа, который не играет основную роль и не связан напрямую с остальными параметрами. То есть мы можем использовать другие данные в этих строчках для анализа, а при анализе рейтинга эти данные всё равно не могут быть учтены. Поэтому заменим значения в % на NaN.

```python
#Заменим значения в % на NaN
data.loc[data['ratings'].str.contains('%') == True, 'ratings'] = np.NaN
```

```python
# check
data['ratings'].unique()
```

```python
#Теперь можем перевести тип данных в float
data['ratings'] = pd.to_numeric(data['ratings'])
```

```python
# check
data.info()
```

Теперь все столбцы имеют правильный тип данных.

#### Шаг 2.2. Изучите пропуски в датафрейме

- Объясните, почему заполнили пропуски определённым образом или почему не стали этого делать.

```python
#посмотрим сколько у нас пропусков
data.isna().sum()
```

```python
#функция для визуализации пропусков
def pass_value_barh(data):
    try:
        (
            (data.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            .plot(kind = 'barh', figsize = (19,6), legend = False, fontsize = 16)
            .set_title('Количество пропусков' + "\n", fontsize = 22, color = 'SteelBlue')    
        );    
    except:
        print('пропусков не осталось :) ')
```

```python
pass_value_barh(data)
```

Столбец `film_studio` нам для дальнейшего анализа не нужен, заполнить их какими-то данными проблематично, да и незачем - оставляем пропуски в таком виде. То же самое можно сказать и про столбец `producer`.

Столбец `box_office` мы тем более не можем заполнить какими-то данными. По кассовым сборам информация либо есть, либо нет, никакую достоверную среднюю из любых данных не высчитать.

Возвратные средства предоставляются реже, чем невозвратные. Мы можем сказать, что суммы при предоставлении возвратных средств примерно такие же, как и при предоставлении невозвратных. Получается, что

Сборы не зависят от бюджета фильма - и недорогой фильм может стать популярным, также сборы не зависят от суммы гос поддержки.    
Всего 27% фильмов с гос финансированием окупают сумму этой гос поддержки. Полностью окупается только 19% фильмов с гос финансированием. Очевидно, что при принятии решения о гос поддержки, фактор окупаемости учитывается далеко не в первую очередь.   


БОльшие суммы гос поддержки выделяются на фильмы:
- которые снимал Бондарчук, Хомерики, Кравчук,
- в жанре драмы, комедии и мультфильмы,
- тип которых анимационный, художественный, документальный, 
- возрастная категория которых 6+ , 12+ и 16+.
